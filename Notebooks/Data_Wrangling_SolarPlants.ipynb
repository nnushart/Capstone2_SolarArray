## 3.0  Data Cleaning

```python
# Check zeros from DC and AC power

zeros_p1dc = p1_gen[p1_gen['DC_POWER'] == 0].agg('count')
total_p1dc = p1_gen['DC_POWER'].agg('count')

zeros_p1ac = p1_gen[p1_gen['AC_POWER'] == 0].agg('count')
total_p1ac = p1_gen['AC_POWER'].agg('count')


zeros_p1dc/total_p1dc, zeros_p1ac/total_p1ac
```

```python
# Check zeros from DC and AC power

zeros_p2dc = p2_gen[p2_gen['DC_POWER'] == 0].agg('count')
total_p2dc = p2_gen['DC_POWER'].agg('count')

zeros_p2ac = p2_gen[p2_gen['AC_POWER'] == 0].agg('count')
total_p2ac = p2_gen['AC_POWER'].agg('count')


zeros_p2dc/total_p2dc, zeros_p2ac/total_p2ac
```

```python
# Check zeros from daily and total yield

zeros_p1dy = p1_gen[p1_gen['DAILY_YIELD'] == 0].agg('count')
zeros_p1ty = p1_gen[p1_gen['TOTAL_YIELD'] == 0].agg('count')

zeros_p1dy/total_p1dc, zeros_p1ty/total_p1dc
```

```python
# Check zeros from daily and total yield

zeros_p2dy = p2_gen[p2_gen['DAILY_YIELD'] == 0].agg('count')
zeros_p2ty = p2_gen[p2_gen['TOTAL_YIELD'] == 0].agg('count')

zeros_p2dy/total_p2dc, zeros_p2ty/total_p2dc
```

#### *Slightly different percentage of zero-values in DC/AC power between the two plants; in plant 2, more than half of the DC & AC power values are zero. Plant 1 has zeros for 27% of daily yield values and Plant 2 has about 17%, this will require investigation in EDA. The good news is that the total yield has zeros for less than 1% of values Plant 2 and Plant 1 doesn't have any zero values.*

```python
# Check the Datetime values match in the plant-specific dataframes

p2_ws['DATE_TIME'].isin(p2_gen['DATE_TIME']).value_counts()
```

```python
# Merge the Plant_1 Dataframes

p1_merged = pd.merge(p1_gen, p1_ws, how='inner', on='DATE_TIME')

p1_merged.head(), p1_merged.shape
```

```python
# Merge the Plant_2 Dataframes

p2_merged = pd.merge(p2_gen, p2_ws, how='inner', on='DATE_TIME')

p2_merged.head(), p2_merged.shape
```

#### *Merged the weather sensor data to the corresponding solar plant. Used "inner" for Plant 1 even though there was a difference in 'DATE_TIME' value counts, it only dropped 4 rows because of null temperature data. Used "inner" for Plant 2 because there were no different values for 'DATE_TIME'.*

#### *Final check on missing values*

```python
# Check missing/null values

missing_p1mrg = pd.concat([p1_merged.isnull().sum(), 100 * p1_merged.isnull().mean()], axis=1)
missing_p1mrg.columns=['count', '%']
missing_p1mrg.sort_values(by='count', ascending=False)
```

```python
# Check missing/null values

missing_p2mrg = pd.concat([p2_merged.isnull().sum(), 100 * p2_merged.isnull().mean()], axis=1)
missing_p2mrg.columns=['count', '%']
missing_p2mrg.sort_values(by='count', ascending=False)
```

```python
# Add column for easier querying by plant

p1_merged['PLANT'] = 'plant_1'
p2_merged['PLANT'] = 'plant_2'
```

```python
# Concatenate the merged dataframes

solar_plants = pd.concat([p1_merged, p2_merged])

solar_plants.sample(20)
```

#### *Concatenated merged dataframes containing generation and weather data per plant into one dataset.*

```python
# Final Check on Merged and Concatenated DF relative to the preliminary dataframes

test_ws = p1_ws[p1_ws['DATE_TIME'] == '2020-05-20 6:00:00']
test_gen = p1_gen[p1_gen['DATE_TIME'] == '2020-05-20 6:00:00']
test_concat = solar_plants.loc[(solar_plants['DATE_TIME'] == '2020-05-20 6:00:00') & (solar_plants['PLANT'] == 'plant_1')]

test_ws, test_gen, test_concat
```
